```python
fig1 = {}

gr = []
for i in experiments:
    print ('X_%s'%(i))
    non_null = data['X_%s'%(i)].notna()
#     print (len(data['X_%s'%(i)][non_null]))
#     print (len(data['Y_%s'%(i)][non_null]))
#     print (len(data['ErrorX_%s'%(i)][non_null]))
#     print (len(data['ErroY_%s'%(i)][non_null]))
    x = pd.to_numeric(data['X_%s'%(i)][non_null], errors='coerce')
    y = pd.to_numeric(data['Y_%s'%(i)][non_null], errors='coerce')
    ex = pd.to_numeric(data['ErrorX_%s'%(i)][non_null], errors='coerce')
    ey = pd.to_numeric(data['ErrorY_%s'%(i)][non_null], errors='coerce')
    #pd.to_numeric(s, errors='coerce')
    temp_gr = rt.TGraphErrors()
    temp_gr.SetName(i)
    temp_gr.GetYaxis().SetRangeUser(0.,0.65)
    temp_gr.GetXaxis().SetRangeUser(0.,6.e3)
#     temp_gr.SetMarkerStyle(rt.kFullCircle)
#     temp_gr.SetMarkerColor(rt.kRed)
#     temp_gr.SetMarkerSize(1.2)
    #print ((x.iloc[0]))

    for j in range(len(x)):
        if(temp_gr.GetName() == 'Chasman'):
            print('%d : %f+/-%f : %f+/-%f '%(j,x.iloc[j],ex.iloc[j],y.iloc[j],ey.iloc[j]))
        temp_gr.SetPoint(j,(x.iloc[j]),y.iloc[j])
        temp_gr.SetPointError(j,ex.iloc[j],ey.iloc[j])
#     print(data['X_%s'%(i)][non_null])
#     print ('Y_%s'%(i))
#     print(data['Y_%s'%(i)][non_null])
    gr.append(temp_gr)


    del temp_gr
    

print("Done!!")


for graph in gr:
    print ('Graph for %s'%(graph.GetName()))

# gr[0].Draw("AP")
# c.Draw()
```

```python
marker_style = [rt.kOpenCross, rt.kFullCircle, rt.kOpenSquare, rt.kOpenDoubleDiamond, rt.kOpenSquare, rt. kMultiply,
                rt.kFullCrossX, rt.kCircle, rt.kOpenTriangleUp, rt.kOpenDiamond, rt.kFullSquare, rt.kPlus, rt.kOpenStar]

marker_color = [rt.kRed, rt.kRed, rt.kGreen, rt.kGreen, rt.kRed,  rt.kRed, rt.kBlue, rt.kGreen, rt.kBlue, rt.kGreen,
                rt.kYellow+1, rt.kBlue, rt.kBlack]
```

```python
c = rt.TCanvas()
count = 0
legend = rt.TLegend(0.12,0.88,0.65,0.65)
legend.SetNColumns(3)
legend.SetBorderSize(0)
for i,graph in enumerate(gr):
#     if(gr[i].GetName() != 'Cogent'):
#         continue
    print ('%d : Graph for %s'%(i,graph.GetName()))
    gr[i].SetMarkerStyle(marker_style[i])
    gr[i].SetMarkerColor(marker_color[i])
    gr[i].SetMarkerSize(1.2)
    gr[i].SetLineColor(marker_color[i])
    
    if(not count):
        gr[i].Draw("APE")
        gr[i].GetYaxis().SetRangeUser(0.,0.65)
        #gr[i].GetXaxis().SetRangeUser(0.5,6.e3)
        gr[i].GetXaxis().SetLimits(0.2,2.e3)
        gr[i].Draw("APE")
        count += 1
    else:
        gr[i].Draw("PE SAME")
        
    l1 = legend.AddEntry(gr[i],gr[i].GetName(),"pe")
    l1.SetTextColor(gr[i].GetMarkerColor())
    
    gr[i].GetXaxis().SetTitle('Nuclear recoil energy [keV]')
    gr[i].GetYaxis().SetTitle('Ionization yield')
    

legend.Draw()
c.Draw()
rt.gPad.SetLogx()
```

```python
YieldModel = pd.read_csv('Fig1_YieldModel_Literature.txt', delimiter='\t',header=None)
YieldModel.columns = [i for i in YieldModel.iloc[0]]
#print(YieldModel.head())
models = ['Lindhard','EDELWEISS17','Scholz','CollarPN']
print(models)
YieldModel = YieldModel.drop([0])
(YieldModel.head())
```

```python
md = []
for j,i in enumerate(models):
    print ('%s'%(i))
    non_null = YieldModel['X_%s'%(i)].notna()
    x = pd.to_numeric(YieldModel['X_%s'%(i)][non_null], errors='coerce')
    y = pd.to_numeric(YieldModel['Y_%s'%(i)][non_null], errors='coerce')
    exl = pd.to_numeric(np.zeros(len(x)))
    exh = pd.to_numeric(np.zeros(len(x)))
    eyl = pd.to_numeric(np.zeros(len(y)))
    eyh = pd.to_numeric(np.zeros(len(y)))
    if(j > 1):
        exl = pd.to_numeric(YieldModel['X_%sLower'%(i)][non_null], errors='coerce')
        exh = pd.to_numeric(YieldModel['X_%sUpper'%(i)][non_null], errors='coerce')
        eyl = pd.to_numeric(YieldModel['Y_%sLower'%(i)][non_null], errors='coerce')
        eyh = pd.to_numeric(YieldModel['Y_%sUpper'%(i)][non_null], errors='coerce')

    temp_gr = rt.TGraphAsymmErrors()#(len(x),x,y,exl,exh,eyl,eyh)
    for k in range(len(x)):
        temp_gr.SetPoint(k,(x.iloc[k]),y.iloc[k])
        if(j > 1):
            temp_gr.SetPointError(k,x.iloc[k]-exl.iloc[k],exh.iloc[k]-x.iloc[k],y.iloc[k]-eyl.iloc[k],eyh.iloc[k]-y.iloc[k])
            if(j == 3):
                print('%d, %0.2f, %0.2f, %0.2f, %0.2f'%(k,exl.iloc[k],exh.iloc[k],eyl.iloc[k],eyh.iloc[k]))
        else:
            temp_gr.SetPointError(k,exl[k],exh[k],eyl[k],eyh[k])

    
    
    temp_gr.SetName(i)
    md.append(temp_gr)


    del temp_gr
#     if(j > 0):
#         break

print("Done!!")


for model_gr in md:
    print ('Graph for %s'%(model_gr.GetName()))
```
